```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
```

```python
df = pd.read_csv('Stroke_Dataset.csv')
```

```python
df
```

```python
df = df.drop(['id'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
missing = 28.893237
df['bmi']=df.bmi.fillna(missing)
```

```python
df.isnull().sum()
```

```python
df['gender'] = df['gender'].replace({'Male':1,'Female':2,'Other':3})
```

```python
df['ever_married'] = df['ever_married'].replace({'Yes':0,'No':1})
```

```python
df = df.drop(['work_type'],axis=1)
```

```python
df = df.drop(['Residence_type'],axis=1)
```

```python
df['smoking_status'] = df['smoking_status'].replace({'Unknown':0.5,'smokes':1,'never smoked':0,'formerly smoked':0.7})
```

```python
df
```

```python
x = df.drop(['stroke'],axis=1)
```

```python
x
```

```python
y = df['stroke']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
xtrain
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(xtrain,ytrain)
```

```python
accuracy = model.score(xtest,ytest)
print("SVM Accuracy : ",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pre = model.predict(xtest)
```

```python
cm = confusion_matrix(ytest,y_pre)
```

```python
plt.figure(figsize=(6,4))
seaborn.heatmap(cm,annot=True,fmt='d',cmap='Blues',xticklabels=['No Stroke','Stroke'],yticklabels=['No Stroke','Stroke'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('SVM Confusion Matrix')
plt.show()
```
